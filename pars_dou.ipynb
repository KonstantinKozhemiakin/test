```python
import requests
from bs4 import BeautifulSoup

# URL of the page with the vacancies
url = "https://jobs.dou.ua/vacancies/?search=lead%20generation&exp=1-3"

# Send a GET request to the URL and retrieve the page content
# response = requests.get(url)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"}
response = requests.get(url, headers=headers)

content = response.content
print(response)
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(content, "html.parser")
# print(soup)
# Find all the vacancy elements on the page
vacancy_elements = soup.find_all("li", class_="l-vacancy")
# print(vacancy_elements)

# Iterate over each vacancy element and extract the necessary information
# ...
vacancy_list =[[],[]]

for vacancy in vacancy_elements:
    # Extract the vacancy URL
    vacancy_url = vacancy.find("a", class_="vt").get("href")
    vacancy_response = requests.get(vacancy_url, headers=headers)
    vacancy_content = vacancy_response.content
    # print(vacancy_content)
    vacancy_soup = BeautifulSoup(vacancy_content, "html.parser")

    title = vacancy_soup.find("h1", class_="g-h2").text.strip()

    try:
        description = vacancy_soup.find("div", class_="text b-typo vacancy-section").text.strip()
        print(description)
    except AttributeError:
        description = "Description not found"

    # Print the extracted information for each vacancy
    print("Title:", title)    
    print("Description:", description)
    print("------------------------------")
    vacancy_list[0].append(title)
    vacancy_list[1].append(description)

# <div class="text b-typo vacancy-section">
```
